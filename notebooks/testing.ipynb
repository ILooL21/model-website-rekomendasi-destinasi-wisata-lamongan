```python
import numpy as np
import pandas as pd
import seaborn as sns
import bnlearn as bn
import pprint as pp
```

```python
model = bn.bnlearn.load(filepath="../model/model.pkl")
```

```python
# Mapping label ke nama wisata
LABEL_DESTINASI = {
    10: 'Pantai Joko Mursodo',
    12: 'Pantai Pengkolan Kandang Semangkon',
    11: 'Pantai Kutang',
    14: 'Taman Wisata dan Perkemahan Bumi Moronyamplung',
    2: 'Gunung Mas Mantup',
    1: 'Goa Maharani',
    19: 'Wisata Gunung Pegat',
    16: 'Wisata Air Hangat Mbrumbung',
    13: 'Taman Ekspresi Kendalifornia',
    15: 'Waduk Gondang',
    4: 'Kolam Renang Keraton',
    18: 'Wisata Edukasi Gondang Outbond',
    3: 'INDONESIA ISLAMIC ART MUSEUM',
    17: 'Wisata Bahari Lamongan',
    0: 'G-Park',
    5: 'Makam Sendang Duwur',
    6: 'Makam Sunan Drajat',
    9: 'Masjid Namira',
    8: 'Masjid Ki Bagus Hadikusumo',
    7: 'Makam Syekh Maulana Ishaq'
}
```

```python
# ambil data pantai joko mursodo
data1 = pd.read_csv('../data/processed/pantai joko mursodo.csv')

random_row1 = data1.sample(n=1).iloc[0]

evidence1 = {
    'escape': int(random_row1['escape']),
    'relaxation': int(random_row1['relaxation']),
    'play': int(random_row1['play']),
    'strengthening_family_bonds': int(random_row1['strengthening_family_bonds']),
    'prestige': int(random_row1['prestige']),
    'social_interaction': int(random_row1['social_interaction']),
    'romance': int(random_row1['romance']),
    'educational_opportunity': int(random_row1['educational_opportunity']),
    'self_fulfillment': int(random_row1['self_fulfillment']),
    'wish_fulfillment': int(random_row1['wish_fulfillment']),
    'infrastruktur_pariwisata': int(random_row1['infrastruktur_pariwisata']),
    'edukasi': int(random_row1['edukasi']),
    'alam_sekitar': int(random_row1['alam_sekitar']),
    'atraksi_budaya_dan_sejarah': int(random_row1['atraksi_budaya_dan_sejarah']),
    'kuliner': int(random_row1['kuliner']),
    'berendam': int(random_row1['berendam']),
    'bermain_wahana': int(random_row1['bermain_wahana']),
    'melihat_pemandangan': int(random_row1['melihat_pemandangan']),
    'mengambil_foto': int(random_row1['mengambil_foto']),
    'berkemah': int(random_row1['berkemah']),
    'berbelanja': int(random_row1['berbelanja']),
    'berziarah': int(random_row1['berziarah']),
    'memancing': int(random_row1['memancing']),
    'berinteraksi_dengan_satwa': int(random_row1['berinteraksi_dengan_satwa']),
    'makan': int(random_row1['makan']),
    'belajar': int(random_row1['belajar']),
    'berenang': int(random_row1['berenang'])
}

print(evidence1)

prediction1 = bn.inference.fit(
    model,
    variables=['Nama Wisata'],
    evidence=evidence1,
    verbose=3,
    elimination_order="MinWeight"
)

prediction1.df['Nama Wisata'] = LABEL_DESTINASI
prediction1.df = prediction1.df.sort_values(by='p', ascending=False)
prediction1.df
```
